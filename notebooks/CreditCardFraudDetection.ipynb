```python
#Segreating numerical and categorical variables
categorical_cols = X.select_dtypes(include='object').columns
numerical_cols = X.select_dtypes(exclude='object').columns
```

```python
from sklearn.impute import SimpleImputer  #Missing values
from sklearn.preprocessing import StandardScaler #Handling Feature Scaling
from sklearn.preprocessing import OrdinalEncoder #ordinal Encoding

#Pipelines
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
#Numerical Pipelines
num_pipeline = Pipeline(
    steps =[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler',StandardScaler())
    ]
)
```

```python
preprocessor = ColumnTransformer([
    ('num_pipeline',num_pipeline,numerical_cols),
])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=0)
```

```python


import seaborn as sns
#Using Pearson Correlation
plt.figure(figsize=(12,10))
cor = X_train.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.CMRmap_r)
plt.show()

```

```python
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature

def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(X_train, 0.9)
len(set(corr_features))
```

```python
corr_features
```

```python
X_train.drop(corr_features,axis=1)
X_test.drop(corr_features,axis=1)
```

```python
X_train=pd.DataFrame(preprocessor.fit_transform(X_train),columns=preprocessor.get_feature_names_out())
X_test=pd.DataFrame(preprocessor.transform(X_test),columns=preprocessor.get_feature_names_out())
```

```python
X_train.head()
```

```python
#  Model Training
from sklearn.naive_bayes  import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, fbeta_score
from sklearn.model_selection import GridSearchCV

```

```python
import numpy as np
def evaluate_model(true,predicated):
    accuracy = accuracy_score(true,predicated)
    recall = recall_score(true, predicated)
    precision = precision_score(true,predicated)
    fbeta = fbeta_score(true, predicated, beta=1)
    return accuracy,recall,precision,fbeta
```

```python
## Train multiple models
## Model Evaluations
models = {
    'GaussianNB' :GaussianNB(),
    'XGBoost' : GradientBoostingClassifier(),
    'RandomForest': RandomForestClassifier(),
    'LogisticRegression': LogisticRegression(),
    'SupportVector': SVC()
}
# for gridsearch cv
param_grid = {
    'GaussianNB': {},
    'XGBoost': {'learning_rate': [0.01, 0.1], 'n_estimators': [100, 500]},
    'RandomForest': {'n_estimators': [100, 500], 'max_depth': [None, 5, 10]},
    'LogisticRegression': {'C': [0.1, 1, 10]},
    'SupportVector': {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
}

trained_model_list = []
model_list=[]
r2_list=[]

for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(X_train,y_train)

    #for grid search cv
    model_name = list(models.keys())[i]
    
    grid_search = GridSearchCV(model, param_grid[model_name], cv=5)
    grid_search.fit(X_train, y_train)
    
    best_model = grid_search.best_estimator_    

    #Make Predictions
    y_pred=model.predict(X_test)
    
    ### if need to apply gridsearch cv use below code
    # y_pred = best_model.predict(X_test)

    accuracy, recall, precision,fbeta=evaluate_model(y_test,y_pred)

    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])

    print('Model Training Performance')
    print("Accuracy score:",accuracy*100)
    print("Recall score:",recall*100)
    print("Precision score:",precision*100)
    print("Fbeta_score :",fbeta*100)

    r2_list.append(accuracy)
    
    print('='*35)
    print('\n')
```
